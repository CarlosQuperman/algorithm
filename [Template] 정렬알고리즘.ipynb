```python
import random
import time

a=random.sample(range(1,1001),1000) # 1~1000의 값을 무작위로 배치한 배열
b=a[:]
c=a[:]
# 만들어진 배열을 3개로 복사한다.
print("Array Size : ",len(a)) # 배열의 크기를 출력한다.
print("NonSotred Array : ",a) # 정렬되지 않은 배열을 출력한다.

start=time.time()
bubbleSort(a) # 버블 정렬을 실행한다.
print("BubbleSort Time : ",time.time()-start,"s") # 버블 정렬 실행 시간
print("BubbleSorted : ",a) # 버블 정렬된 결과를 출력한다.

start=time.time()
selectionSort(b) # 선택 정렬을 실행한다.
print("SelectionSort Time : ",time.time()-start,"s") # 선택 정렬 실행 시간
print("SelectionSotred : ",b) # 선택 정렬된 결과를 출력한다.

start=time.time()
insertionSort(c) # 삽입 정렬을 실행한다.
print("InsertionSort Time : ",time.time()-start,"s") # 삽입 정렬 실행 시간
print("InsertionSorted : ",c) # 삽입 정렬된 결과를 출력한다.
```

```python
def mergeSort(x): # 합병 정렬
    if len(x) > 1:
        mid = len(x)//2
        lx, rx = x[:mid], x[mid:] # 가운데를 중심으로 배열을 2개로 나눈다,
        mergeSort(lx) # 왼쪽 배열을 합병 절렬한다.
        mergeSort(rx) # 오른쪽 배열을 합병 정렬한다.

        li, ri, i = 0, 0, 0
        while li < len(lx) and ri < len(rx): # 왼쪽 배열과 오른쪽 배열의 앞의 값들을 비교해서
            if lx[li] < rx[ri]: # 왼쪽이 작으면 왼쪽 배열의 값을 넣고 위치를 한 칸 뒤로 하고
                x[i] = lx[li]
                li += 1
            else: # 아니면 오른쪽 배열의 값을 넣고 위치를 한 칸 뒤로 한다.
                x[i] = rx[ri]
                ri += 1
            i += 1
        x[i:] = lx[li:] if li != len(lx) else rx[ri:] # 둘 배열중 한쪽이 다 끝나면 남은 배열의 나머지를 넣는다.
```

```python
def pivotFirst(x, lmark, rmark): # 퀵 정렬
    pivot = x[lmark] # 기준 값을 설정한다.
    pivot_i = lmark # 기준 값의 위치
    while lmark <= rmark: # 기준 값을 기준으로 작은 값은 왼쪽, 큰 값은 오른쪽에 넣는다.
        while lmark <= rmark and x[lmark] <= pivot:
            lmark += 1
        while lmark <= rmark and x[rmark] >= pivot:
            rmark -= 1
        if lmark <= rmark:
            x[lmark],x[rmark]=x[rmark],x[lmark]
            lmark += 1
            rmark -= 1
    x[pivot_i],x[rmark]=x[rmark],x[pivot_i]
    return rmark

def quickSort(x): # 퀵 정렬
    def _qsort(x, first, last):
        if first < last:
            splitpoint = pivotFirst(x, first, last) # 기준 값을 기준으로 작은 값은 왼쪽, 큰 값은 오른쪽에 넣는다.
            _qsort(x, first, splitpoint-1)
            _qsort(x, splitpoint+1, last) #기준 값을 기준으로 나눈 두 배열도 퀵 정렬은 한다.
    _qsort(x, 0, len(x)-1)
```

```python
def heapInput(heap, i, n): # 최댓 힙을 사용하면 힙의 맨 앞에는 가장 큰 값이 저장된다.
    l = i*2;
    r = i*2+1
    s_i = i
    if (l <= n and heap[s_i] < heap[l]):
        s_i = l
    if (r <= n and heap[s_i] < heap[r]):
        s_i = r
    if s_i != i:
        heap[i], heap[s_i] = heap[s_i], heap[i]
        heapInput(heap, s_i, n)
 
def heapSort(x): # 힙 정렬
    n = len(x)
    heap = [0]+x
 
    for i in range(n, 0, -1): # 배열에 있는 값을 모두 최댓 힙에 넣는다.
      heapInput(heap, i, n)
 
    for i in range(n, 0, -1): # 최댓 힙에서 값을 추출해 맨 뒤에 넣는 것을 반복해서 정렬한다.
      x[i-1]=heap[1]
      heap[i], heap[1] = heap[1], heap[i]
      heapInput(heap, 1, i-1)
```
